```python
directorySource = Path(r'C:\Users\AG(A&E)\MKT_Workstation\Personal_Projects\SAS Quizzer Web App\data_back\pointers')

directoryDestination = Path(r'C:\Users\AG(A&E)\MKT_Workstation\Personal_Projects\sasquizzer2025\data\pointers')

def copy_files_from_csv(file: Path, source: Path, destination: Path):
    # Read CSV with pandas
    df = pd.read_csv(file)
    
    # Extract the 'code quizefile_old' column (assuming exact column name)
    files_to_copy = df['quizefile_old'].dropna().unique()
    
    # Ensure destination folder exists
    destination.mkdir(parents=True, exist_ok=True)
    
    for filename in files_to_copy:
        filename =f'{filename}.csv'
        source_file = source / filename
        if source_file.exists() and source_file.is_file():
            shutil.copy2(source_file, destination / filename)
            print(f"File copied succesfully : {source_file}")
        else:
            print(f"File not found: {source_file}")
copy_files_from_csv(filePath, directorySource , directoryDestination)
```

```python

def renameFiles(filePath, directory):
    df = pd.read_csv(filePath)  
    for _, row in df.iterrows():
        old_file = directory / f'{row['quizefile']}.csv'
        new_file = directory / f'{row['quizefile_new']}.csv'
        
        # Check if the old file exists, then rename
        if old_file.exists():
            old_file.rename(new_file)
            print(f"Renamed '{old_file.name}' to '{new_file.name}'")
        else:
            print(f"File '{old_file.name}' not found.")


```
